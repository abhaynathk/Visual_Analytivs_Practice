```python
import requests
from bs4 import BeautifulSoup

# The URL of the website you want to scrape
url = 'https://science.nasa.gov/exoplanets/exoplanet-catalog/'

# Send a request to fetch the web page
response = requests.get(url)

# Parse the content of the page with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract descriptive data, e.g., all paragraph tags
paragraphs = soup.find_all('p')

# Loop through each paragraph and print the text
for paragraph in paragraphs:
    print(paragraph.get_text())

# You can also extract headings, such as h1, h2, etc.
headings = soup.find_all(['h1', 'h2', 'h3'])
for heading in headings:
    print(heading.get_text())
```

```python
data1 = {
    'pl_orbper': np.random.lognormal(mean=1, sigma=1, size=100),
    'pl_bmasse': np.random.lognormal(mean=1, sigma=1, size=100),
    'method': np.random.choice(['Transit', 'Radial Velocity', 'Imaging', 'Microlensing', 'Other'], 100)
}
```

```python
df2 = pd.DataFrame(data1)
```

```python
color_map = {
    'Transit': 'yellow',
    'Radial Velocity': 'blue',
    'Imaging': 'red',
    'Microlensing': 'brown',
    'Other': 'gray'
}
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('black')
fig.patch.set_facecolor('black')
# Plot each method in a loop
for method, color in color_map.items():
    filtered_data = df2[df2['method'] == method]
    ax.scatter(
        filtered_data['pl_orbper'],
        filtered_data['pl_bmasse'],
        color=color,
        label=method,
        edgecolor='k',
        alpha=0.6
    )
# Set the scale to logarithmic for both axes
ax.set_xscale('log')
ax.set_yscale('log')

# Add labels and title
ax.set_xlabel('Orbit Period (Earth Days)', color='white')
ax.set_ylabel('Planet Mass (Earth Masses)', color='white')
ax.set_title('Exoplanet Census', color='white')

# Set grid color and style


# Customize tick colors
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Add legend
legend = ax.legend(facecolor='black', edgecolor='white', loc='upper right')
for text in legend.get_texts():
    text.set_color('white')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the categories
labels = ['Neptune-like', 'Gas Giant', 'Super Earth', 'Terrestrial', 'Unknown']
sizes = [1961, 1850, 1733, 205, 7]
colors = ['#FF4500', '#FF6347', '#FF7F50', '#FF8C00', '#FF4500']  # shades of red-orange

# Create figure and axis with a dark background
fig, ax = plt.subplots(figsize=(20, 13), subplot_kw=dict(aspect="equal"))
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Pie chart, turn it into a donut by setting wedgeprops width
wedges, texts = ax.pie(
    sizes, 
    labels=None, 
    startangle=140, 
    colors=colors, 
    wedgeprops=dict(width=0.4),  # control the width to make a donut shape
    autopct=None  # No text inside the wedges
)

# Add central text for total exoplanets
ax.text(0, 0, '5756\nCONFIRMED\nEXOPLANETS', 
        ha='center', va='center', 
        fontsize=24, color='white', fontweight='bold')

# Add the labels and values next to the arcs
for i, (label, size) in enumerate(zip(labels, sizes)):
    angle = wedges[i].theta2 - (wedges[i].theta2 - wedges[i].theta1) / 2  # calculate angle
    x = wedges[i].r * 1.2 * np.cos(np.radians(angle))  # position x slightly away
    y = wedges[i].r * 1.2 * np.sin(np.radians(angle))  # position y slightly away
    ax.text(x, y, f"{size}\n{label}", ha='center', va='center', color='white', fontsize=14)

# Hide the axes for clean appearance
plt.axis('off')

# Display the plot
plt.show()
```
